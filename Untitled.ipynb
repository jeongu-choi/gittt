```python
import sys
import pandas as pd
from PySide6.QtWidgets import QApplication, QMainWindow, QTableWidgetItem, QWidget
from yoni import Ui_MainWindow  # 생성된 파이썬 파일의 이름에 따라 import 구문이 달라질 수 있습니다.
from secondwindow import Ui_Form

#window_2 = None
class MainWindow(QMainWindow):
    def __init__(self):
        super().__init__()
        self.ui = Ui_MainWindow()
        self.ui.setupUi(self)
        self.ui.H_beam.clicked.connect(self.groupboxRadFunction)
        self.ui.I_beam.clicked.connect(self.groupboxRadFunction)
        self.ui.Bridge_Length.returnPressed.connect(self.print_Bridge_Length)
        self.window_2 = None

    def groupboxRadFunction(self):
        if self.ui.H_beam.isChecked():
            window_2 = TableWidgetDemo()
            window_2.show()

        elif self.ui.I_beam.isChecked():
            print("I_beam Checked")

    def print_Bridge_Length(self):
        # Lineedit에 있는 글자를 가져오는 메서드
        print(self.ui.Bridge_Length.text())


class TableWidgetDemo(QWidget):
    def __init__(self):
        super(TableWidgetDemo, self).__init__()
        self.ui = Ui_Form()
        self.ui.setupUi(self)

        # CSV 파일 읽기
        self.load_csv_data("your_excel_file.csv")

    def load_csv_data(self, file_path):
        # pandas를 사용하여 CSV 파일 읽기
        df = pd.read_csv(file_path)

        # 행과 열의 수 설정
        self.ui.tableWidget.setRowCount(len(df))
        self.ui.tableWidget.setColumnCount(len(df.columns))
        self.ui.tableWidget.setHorizontalHeaderLabels(df.columns)

        # 데이터 추가
        for row in df.itertuples():
            for col_index, value in enumerate(row[1:], start=0):  # itertuples의 첫 번째 값은 인덱스이므로 제외
                self.ui.tableWidget.setItem(row.Index, col_index, QTableWidgetItem(str(value)))


if __name__ == "__main__":
    app = QApplication(sys.argv)
    window = MainWindow()
    window.show()
    sys.exit(app.exec())
```
